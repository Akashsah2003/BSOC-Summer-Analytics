```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
df1 = pd.read_csv('world-happiness-report-2021.csv')
```

```python
df1.head(5)
```

```python
df1.tail(10)
```

```python
df1.info()
```

```python
fig=plt.figure(figsize=(25,5))
plt.subplot(1, 3, 1)
sns.kdeplot(df1['Ladder score'], shade=True)
plt.subplot(1, 3, 2)
sns.histplot(x=df1['Ladder score'], kde=True)
plt.subplot(1, 3, 3)
sns.boxplot(x=df1['Ladder score'])
print(df1['Ladder score'].describe())
# sns.histplot(x=df1['Ladder score'], kde=True)
```

# Ladder score - 1 outlier

```python
fig=plt.figure(figsize=(25,5))
plt.subplot(1, 3, 1)
sns.kdeplot(df1['Standard error of ladder score'], shade=True)
plt.subplot(1, 3, 2)
sns.histplot(x=df1['Standard error of ladder score'], kde=True)
plt.subplot(1, 3, 3)
sns.boxplot(x=df1['Standard error of ladder score'])
print(df1['Standard error of ladder score'].describe())
```

# Standard error of ladder score - skewed, 3 outliers

```python
fig=plt.figure(figsize=(25,5))
plt.subplot(1, 3, 1)
sns.kdeplot(df1['upperwhisker'], shade=True)
plt.subplot(1, 3, 2)
sns.histplot(x=df1['upperwhisker'], kde=True)
plt.subplot(1, 3, 3)
sns.boxplot(x=df1['upperwhisker'])
print(df1['upperwhisker'].describe())
```

```python
fig=plt.figure(figsize=(25,5))
plt.subplot(1, 3, 1)
sns.kdeplot(df1['Healthy life expectancy'], shade=True)
plt.subplot(1, 3, 2)
sns.histplot(x=df1['Healthy life expectancy'], kde=True)
plt.subplot(1, 3, 3)
sns.boxplot(x=df1['Healthy life expectancy'])
print(df1['Healthy life expectancy'].describe())
```
